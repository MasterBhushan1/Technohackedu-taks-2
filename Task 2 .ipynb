## Calculating Median and Mode

```python
df
```

```python
num_value = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
num_value
```

```python
for i in num_value:
    df_med = df[i].median()
    print(f"median of {i} is", df_med)
    
```

```python
for i in num_value:
    df_mode =df[i].mode()
    print(f"mode of {i} is", df_mode)
    
```

```python
cat_value = df.select_dtypes(include=['object']).columns.tolist()
cat_value
```

```python
for i in cat_value:
    df_mode = df[i]. mode()
    print(f"mode of {i} is", df_mode)
```

## Here, mode of categorical columns is the value is repeated maximum number of times.
